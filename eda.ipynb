```python
import nibabel as nib
import numpy as np

spacings = []
for p in TRAIN_IMAGE_PATHS:
    img = nib.load(p)
    # The header holds the voxel dimensions (usually indices 1, 2, 3)
    spacings.append(img.header.get_zooms())

print("Median spacing:", np.median(spacings, axis=0))
print("Min spacing:", np.min(spacings, axis=0))
print("Max spacing:", np.max(spacings, axis=0))
print("Mean spacing:", np.mean(spacings, axis=0))
print("Spacing std:", np.std(spacings, axis=0))
```
